```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import re
```

```python
df = pd.read_csv("../output/project_topics_stacked.csv", sep=";", decimal=",", index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df_grouped = df.groupby(["topic", "fp", "fp_no"]).sum().drop(columns=["rcn"]).reset_index()
df_grouped.head()
```

```python
def normalize_load(row):

    fp_no = row["fp_no"]
    load = row["load"]
    
    switcher = {
        0: 2714,
        1: 2692,
        2: 4180,
        3: 11000,
        4: 12000,
        5: 9971,
        6: 26000,
        7: 26000,
    }
    
    denominator = switcher.get(fp_no, "Invalid framework programme number")
    
    return load/denominator    

normalized_loads = df_grouped.apply(lambda row: normalize_load(row), axis=1)
df_grouped["normalized_load"] = normalized_loads
```

```python
def remove_project_suffix(text):
    return re.sub("_projects", "", text).upper()

fp_clean = df_grouped["fp"].apply(lambda text: remove_project_suffix(text))
df_grouped["fp_clean"] = fp_clean
```

```python
df_grouped
```

```python
def plot_topic_dynamics(ax, df_grouped, topic, topic_label):

    df_topic = df_grouped[df_grouped.topic == topic].reset_index(drop=True)
    
    # Plot topic dynamics
    ax.plot(df_topic["fp_clean"], df_topic["normalized_load"], label = f"Topic {topic}", linestyle="-", color="black")
    
    # Plot average topic weight for rerfence
    ax.plot(df_topic["fp_clean"], [0.05]*8, label = "Average Weight", linestyle="--", color="black")
    
    # Set limits
    ax.set_ylim([0,0.15])
    
    # Set title and labels
    ax.set(title=f"Dynamics for Topic {topic} '{topic_label}'", xlabel='Framework Programme', ylabel='Relative Topic Weight')
    ax.legend()
```

```python
topics = [2, 17]
topic_labels = ["Information Systems", "Energy Innovations"]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=[15, 5])
plt.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95, wspace=0.3, hspace=0.3)

for i in range(len(topics)):
    plot_topic_dynamics(axs[i], df_grouped, topics[i], topic_labels[i])
```
