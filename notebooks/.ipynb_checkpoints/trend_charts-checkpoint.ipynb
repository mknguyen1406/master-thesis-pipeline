```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import re
```

```python
first_topic = 2
second_topic = 17
topic_labels = [
        'Agricultural Science',
        'Energy Science',
        'Information Systems',
        'Chemistry',
        'Aerodynamics',
        'Cell Research',
        'Research Programmes',
        'Material Science',
        'Astrophysics',
        'European Development',
        'Social Science',
        'Health Science',
        'Genetic Research',
        'Electronics and Photonics',
        'Neuroscience',
        'Quantum Physics',
        'Molecular Biology',
        'Energy Innovations',
        'Software Engineering',
        'Climate Science',
    ]
```

```python
df = pd.read_csv("../output/topics/all_topics.csv", sep=";", decimal=",")
```

```python
df["hotness"] = df["dif_e5"]/(df["load"]*100000)
```

```python
def remove_project_suffix(text):
    return re.sub("_projects", "", text).upper()

df["fp"] = df["time"].apply(lambda text: remove_project_suffix(text))
```

## 1. Hot Top Words 

```python
def get_top_10_hot_words_load_dif(ax, df, time, topic, topic_label, top_n, custom_label_y=None):
    
    # Get subset of data
    df_sub = df[(df.time == time) & (df.topic == topic)]
    df_sub = df_sub.sort_values("hotness", ascending=False).reset_index()[:top_n][["word", "load", "dif_e5"]]
    
    # Get columns
    loads = df_sub.load
    dif = df_sub.dif_e5
    words = df_sub.word
    
#     print(dif)

    # Set limits
    ax.set_xlim([0, df_sub["load"].max()+0.002])
    ax.set_ylim([df_sub["dif_e5"].min()-10,df_sub["dif_e5"].max()+10])
    
    # Set title and labels
    title = f"Trending up - Top {top_n} for Topic {topic} '{topic_label}' in {re.sub('_projects', '', time).upper()}"
    ax.set(title=title, xlabel="Word Probability", ylabel="Word Probability Difference (x10^5)")

    for i,word in enumerate(words):
        x = loads[i]
        y = dif[i]
        
        if custom_label_y:
            y_label = custom_label_y[i]
            ax.scatter(x, y, marker='o', color='black')
            ax.text(x+0.00015, y_label, word, fontsize=10)
        else:
            ax.scatter(x, y, marker='o', color='black')
            ax.text(x+0.0002, y, word, fontsize=10)
```

## 2. Hot Top Words by Hotness Score

```python
def get_top_10_hot_words(ax, df, time, topic, topic_label, top_n, custom_label_y=None, custom_label_x=None):
    
    # Get subset of data
    df_sub = df[(df.time == time) & (df.topic == topic)]
    df_sub = df_sub.sort_values("hotness", ascending=False).reset_index()[:top_n][["word", "load", "hotness"]]
    
    # Get columns
    loads = df_sub.load
    hotness = df_sub.hotness
    words = df_sub.word
    
#     print(loads)
    
    # Set limits
    x_lower_lim = min(0, df_sub["load"].min()-0.0012)
    ax.set_xlim([x_lower_lim, df_sub["load"].max()+0.002])
    ax.set_ylim([df_sub["hotness"].min()-0.02,df_sub["hotness"].max()+0.02])
    
    # Set title and labels
    title = f"Trending up - Top {top_n} for Topic {topic} '{topic_label}' in {re.sub('_projects', '', time).upper()}"
    ax.set(title=title, xlabel="Word Probability", ylabel="Relative Word Probability Increase")

    for i,word in enumerate(words):
        x = loads[i]
        y = hotness[i]
        ax.scatter(x, y, marker='o', color='black')
        
        if custom_label_y:
            y_label = custom_label_y[i]
            if custom_label_x:
                x_label = custom_label_x[i]
                ax.text(x_label+0.00015, y_label, word, fontsize=10)
        else:
            ax.text(x+0.0002, y, word, fontsize=10)
```

```python
def get_top_10_flop_words_load_dif(ax, df, time, topic, topic_label, top_n, custom_label_y=None):
    
    # Get subset of data
    df_sub = df[(df.time == time) & (df.topic == topic)]
    df_sub = df_sub.sort_values("dif_e5", ascending=True).reset_index()[:top_n][["word", "load", "dif_e5"]]

    # Get columns
    loads = df_sub.load
    dif = df_sub.dif_e5
    words = df_sub.word
    
#     print(dif)

    # Set limits
    ax.set_xlim([0, df_sub["load"].max()+0.004])
    ax.set_ylim([df_sub["dif_e5"].min()-10,df_sub["dif_e5"].max()+10])
    
    # Set title and labels
    title = f"Trending down - Top {top_n} for Topic {topic} '{topic_label}' in {re.sub('_projects', '', time).upper()}"
    ax.set(title=title, xlabel="Word Probability", ylabel="Word Probability Difference (x10^5)")

    for i,word in enumerate(words):
        x = loads[i]
        y = dif[i]
        
        if custom_label_y:
            y_label = custom_label_y[i]
            ax.scatter(x, y, marker='o', color='black')
            ax.text(x+0.0004, y_label, word, fontsize=10)
        else:
            ax.scatter(x, y, marker='o', color='black')
            ax.text(x+0.0004, y, word, fontsize=10)
```

## 4. Hot Top Words by Hotness Score Delta

```python
def get_top_10_hot_words_delta(ax, df, time_1, time_0, topic, topic_label, top_n, custom_label_y=None, custom_label_x=None):
    
    # Create sub data frame for time period 1
    df_sub_1 = df[(df.time == time_1) & (df.topic == topic)]
    df_sub_1 = df_sub_1.sort_values("hotness", ascending=False).reset_index()[:top_n][["word", "load", "hotness"]]

    # Sort by word to ensure same order as in time 0
    df_sub_1 = df_sub_1.sort_values("word").reset_index(drop=True)
#     print(df_sub_1.head(10))

    # Create sub data frame for time period 0
    def in_target_words(word, target_words):
        return word in target_words

    # Get target word list from time period 1 to sample word list from time period 0
    target_words = list(df_sub_1["word"])

    # Get all words from time period 0 and respective topic
    df_sub_0 = df[(df.time == time_0) & (df.topic == topic)][["word", "load", "hotness"]]

    # Sample words based on target word list from time period 1
    in_target_words_bool = df_sub_0.word.apply(lambda word: in_target_words(word, target_words))
    df_sub_0 = df_sub_0[in_target_words_bool]

    # Sort by word to ensure same order as in time 1
    df_sub_0 = df_sub_0.sort_values("word").reset_index(drop=True)

    # Get columns
    loads_1 = df_sub_1.load
    loads_0 = df_sub_0.load
    hotness_1 = df_sub_1.hotness
    hotness_0 = df_sub_0.hotness
    words = df_sub_1.word
    
    # Set limits
    # ax.set_xlim([0, df_sub_1["load"].max()+0.002])
    # ax.set_ylim([df_sub_1["hotness"].min()-0.05, df_sub_1["hotness"].max()+0.05])
    x_upper_border = max(df_sub_1["load"].max()+0.001, 0.007)
    ax.set_xlim([0, x_upper_border])
    # ax.set_ylim([df_sub_1["hotness"].min()-0.05, df_sub_1["hotness"].max()+0.05])

    # Set title and labels
    title = f"Trending up - Changes of Top {top_n} for Topic {topic} '{topic_label}' from {re.sub('_projects', '', time_0).upper()} to {re.sub('_projects', '', time_1).upper()}"
    ax.set(title=title, xlabel="Word Probability", ylabel="Relative Word Probability Increase")

    ax.scatter(loads_1, hotness_1, marker='o', color='black', label=time_1)
    ax.scatter(loads_0, hotness_0, marker='o', color='white', label=time_0, edgecolors='black')

    for i, word in enumerate(words):
        x_1 = loads_1[i]
        x_0 = loads_0[i]
        y_1 = hotness_1[i]
        y_0 = hotness_0[i]
       
        # Create arrow
        ax.annotate("",
                xy=(x_1, y_1), xycoords='data',
                xytext=(x_0, y_0), textcoords='data',
                arrowprops=dict(arrowstyle="->", color="0.5", shrinkA=5, shrinkB=5,))
        
        # Add label
        if custom_label_y:
            y_label = custom_label_y[i]
            if custom_label_x:
                x_label = custom_label_x[i]
                ax.text(x_label+0.00005, y_label-0.001, word, fontsize=10)
            else:
                ax.text(x_1+0.00005, y_label+0.001, word, fontsize=10)
#         else:
#             ax.text(x_1+0.0001, y_label, word, fontsize=10) 

    ax.legend()
```

```python
time_0 = "fp7_projects"
time_1 = "h2020_projects"

topic = 2
top_n = 10

custom_label_y_t2 = [
    0.526309,
    0.567379-0.005,
    0.601500+0.005,
    0.672756,
    0.410890-0.008,
    0.431606,
    0.432970,
    0.510636,
    0.564929,
    0.440519+0.005,
]


custom_label_y_t12 = [
0.496046+0.009,
0.498363,
0.429591,
0.380708+0.007,
0.493413-0.009,
0.513368,
0.417275,
0.397313-0.001,
0.464800,
0.493910-0.003,
]


custom_label_x_t12 = [
0.001043-0.0005,
0.002038,
0.000844,
0.000533-0.00048,
0.001644,
0.005446,
0.003028,
0.000777,
0.000835,
0.001081,
]


fig, axs = plt.subplots(nrows=2, ncols=1, figsize=[12, 10])
plt.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95, wspace=0.3, hspace=0.3)

get_top_10_hot_words_delta(axs[0], df, time_1, time_0, first_topic, topic_labels[first_topic], top_n, custom_label_y_t2)
get_top_10_hot_words_delta(axs[1], df, time_1, time_0, second_topic, topic_labels[second_topic], top_n, custom_label_y_t12, custom_label_x_t12)
```
